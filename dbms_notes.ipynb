2 tier vs 3 tier architectures



2 tier architectures:

* It has 2 layers. one is client layer another is server/database layer
* These kind of databases are easy to maintain
* But these are not scalable and has potential security issues due to direct interaction of clients with databases



3 tier architectures

* It introduces a new layer that is application layer or business layer that interfaces both client and server layer. When the client is requesting something it will be processed by the application layer and then it will go to server layer

3 level of abstraction (three schema architecture)




The main motive of this architecture is to abstarct the data from from client so that the client cannot know where is the data stored

Now it has three levels, client, logical and physical level

The client level decides what kind of data the user can query or how they want to show it to users or to clients
THe logical level decides how to design a database as per the requirements from the client side. they decide how many tables will be created, what will be the columns in each table etc
The physical level decides how to store the database that is designed by the logical level

Data independence




This means that changes in one level shouldnot affect the working of another level.
For example if the client level wants to change how the users see the data on UI, this should not change the data base design
If logical level wants to chnage the database design, it should not hamper the client side design and the physical level design
if physical level wants to chnage where or how to store the database,it should not hamper the client side design and the logical level design

THe client level is view level where the data is being viewed, the concepctual schema is how logical level designs the db, the physical schema is how the physical level store and manipulates the db

Integrity Constraints



The integrity constraints help in maintaining the quality of the data by imposing some constraints. They are used to 
1. Maintain data accuracy: Prevent invalid or inappropriate values from being inserted into tables 
2. Maintain data consistency: Establish relationships between tables, ensuring consistency across related data 
3. Enforce business rules: Ensure that data is entered correctly 

Some examples of integrity constraints are:
1. Entity: Unique and non-null primary keys. 
2. Referential: Matching foreign and primary keys
3. Domain: Data type and range. for example age cannot be in negative
4. Key: Unique primary keys. candidate key 



Keys:

Candidate key: a candidate key is a unique identifier for a record in a table that can be chosen as the primary key. candidate key can be any key that can help to differentiate two identical data entries from each other
For example, we have student db where name, age, gender is mentioned. now there can be any two students with same name, age and gender. hence we need a key that can help us to differentiate between these two students. this key is called candidate key. In this example, a candidate key can be addhar number or licenese number or mobile number etc anything that is unique to student
